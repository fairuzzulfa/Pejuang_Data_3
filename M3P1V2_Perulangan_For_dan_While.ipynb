# **Looping / Perulangan**

Python menyediakan dua statement untuk melakukan proses perulangan yaitu for dan while. Diantara kedua statment ini, secara umum for lebih banyak digunakan daripada while.

# For Loop

Perulangan for digunakan untuk melakukan iterasi didalam squence seperti list, tuple, string, dan array atau objek yang dapat diiterasi lainnya.

**For Loop Syntax**



```
for val in sequence:
    loop body
```



**For Loop Flowchart**

for loop fc.PNG

**Example : Python for Loop**

````python
#Program untuk menghitung jumlah semua angka dalam sebuah list
L1 = [2, 4, 5, 7, 9]
jumlah = 0

#iterasi didalam list
for i in L1:
  jumlah = jumlah + i

print(jumlah)
````

**The range() function**

````python
#Program untuk melakukan iterasi didalam list melalui indeks
minuman = ["teh", "boba", "kopi", "susu"]
p = len(minuman)

#iterasi didalam list melalui indeks
for i in range(p):
  print(minuman[i])
````

**for loop with else**

````python
#Contoh for Loop menggunakan else
for i in range(p):
  print(minuman[i])
else:
  print("Elemen sudah habis")
````

**Latihan 1**

Diberikan list_tagihan untuk menghitung jumlah seluruh tagihan. Gunakan looping for untuk menyelesaikan tugas ini! simpan hasilnya pada variabel total_tagihan!



```
list_tagihan = [50000, 75000, -150000, 125000, 300000, -50000, 200000]
```



````python
list_tagihan = [50000, 75000, -150000, 125000, 300000, -50000, 200000]
total_tagihan = 0

for i in list_tagihan:
  total_tagihan = total_tagihan + i

print(total_tagihan)
````

# While Loop

While loop pada Python digunakan untuk melakukan iterasi pada blok kode selama nilai dari test expression adalah true. 

**Syntax of while Loop in Python**



```
while test_expression:
  Body of while
```



**While Loop Flowchart**

while loop fc.PNG

**Example : Python while Loop**

````python
#Program untuk menjumlahkan semua
#bilangan asli hingga suku ke-n
#jum = 1+2+3+...+n

n = 10
jumlah = 0
i = 1

while i <= n:
  jumlah = jumlah + i
  i = i + 1

print(jumlah)
````

**While loop with else**

Sama dengan for loop kita juga dapat mengkombinasikan while loop dengan blok else. Bagian else akan dieksekusi jika nilai pada test expression pada while loop bernilai false. Perhatikan contoh berikut:

````python
jum = 0
i = 1
while i <= n:
  jum = jum + i
  i = i + 1
else:
  print(jum) #dieksekusi diluar looping
````

**Latihan 2**

Diberikan list tagihan hitunglah total_tagihan menggunakan while loop.



```
tagihan = [50000, 75000, 125000, 300000, 200000]
```



````python
tagihan = [50000, 75000, 125000, 300000, 200000]
jumlah = len(tagihan)
i = 0
total_tagihan = 0

while i < jumlah :
  total_tagihan = total_tagihan + tagihan[i]
  i = i + 1

print(total_tagihan)
````
